```python
# Standard Library Imports
from pathlib import Path

# Installed packages
import pandas as pd

# Our package
from ydata_profiling import ProfileReport
from ydata_profiling.utils.cache import cache_file
```

**Malaysian Dataset and Data Preparation**

We use the same dataset from case study 1 that has been cleaned in data preparation process which is Property Listing in Kuala Lumpur

**Data Preparation**

Start profiling by loading pandas DataFrame.

```python
# Read the KL Properties Dataset that has been cleaned
file_name = cache_file(
    "cleaned_KLProperties.csv",
    "https://raw.githubusercontent.com/drshahizan/Python_EDA/main/assignment/bdm/KicapSambal/cleaned_KLProperties.csv",
)
df = pd.read_csv(file_name)
```

**Implementation of Automated EDA Tools**

Generate the standard profiling report to see the analysis of data from Pandas Profiling

```python
# Generate the Profiling Report
profile = ProfileReport(
    df, title="KLProperties_Dataset", html={"style": {"full_width": True}}, sort=None
)
```

```python
# The HTML report in an iframe
profile
```

To generate a HTML report file, save the Profile Report to an object and use the to_file() function

```python
profile.to_file("KLProperties_Report.html")
```
